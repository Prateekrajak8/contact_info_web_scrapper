```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
df = pd.read_csv('input.csv')
driver = webdriver.Chrome()
url_list = df.iloc[:, 1].tolist()
address_list = df.iloc[:, 0].tolist()
results = []

for i,url in enumerate(url_list):
    account = address_list[i]
    try:
        driver.get(url)
        driver.implicitly_wait(10)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        contact_us_info = soup.find('p', class_='showcase-address tc')
        contact_url = soup.find('ul', class_='showcase-web-phone ml0 mb3 list tc')
        contact_person = soup.find('div', class_='grid grid-3-col')
        services = []
        if contact_us_info is None:
            raise ValueError("Contact us info not found")
        contact_us_info = contact_us_info.text.strip()    
        result = {'Account': account,'URL': url,'Contact address': contact_us_info}
        for li in contact_url.find_all('li'):
            span = li.find('span')
            if span is not None:
                key = span.text.strip().rstrip(':')
                value = li.contents[-1].strip()
            else:
                key = 'Link'
                value = li.text.strip()
            result[key] = value
        if contact_person is not None:  
            contact_divs = soup.find_all('div', class_='vbc_contact')
            for i, div in enumerate(contact_divs, start=1):
                name = div.find('h3', class_='dib f2 ma0 mb1')
                designation = div.find('p', class_='f3 lh-title muted tc')
                content = div.find('p', class_='description')
                if name:
                    key = f'Name{i}'
                    value = name.text.strip()
                    result[key] = value  
                if designation:
                    key = f'Designation{i}'
                    value = designation.text.strip()
                    result[key] = value  
                if content:
                    key = f'Description{i}'
                    value = content.text.strip()  
                    result[key] = value
        results.append(result)
    except Exception as e:
        print(f"Error processing URL {url}: {str(e)}")

output_df = pd.DataFrame(results)

output_df.to_csv('output.csv', index=False)
```
